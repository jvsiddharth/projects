```python
#Cleaning dataset to find correlation matrix
```

```python
print(df.isnull().sum())
df = df.dropna()
```

```python
from scipy import stats

z_scores = stats.zscore(df['Gross'])
threshold = 3  # Adjust the threshold based on your data
outliers = df['Gross'][np.abs(z_scores) > threshold]

# Remove outliers
df = df[~df['Gross'].isin(outliers)]
```

```python
df['Title'] = pd.to_numeric(df['Title'], errors='coerce')
numeric_df = df.select_dtypes(include=['number'])
```

```python
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
